```python
# Imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Draw plots inline
%matplotlib inline
```

```python
df = pd.read_csv('data/data/USA.csv')
```

```python
df.info()
```

```python
fig, ax = plt.subplots()

# City Count
x = df.url_name_x.value_counts()

ax.bar(x.index, x.values )
ax.set_title('USA Transit System City Frequency')
ax.set_xlabel('City')
ax.set_ylabel('Frequency')
plt.xticks(rotation=90)
```

```python
#  
x = df.pivot_table(index = 'url_name_x', values = 'name_y', aggfunc = len).reset_index()

fig, ax = plt.subplots()

ax.bar(x['url_name_x'], x['name_y'] )
ax.set_title('Number of Rail Lines Per City')
ax.set_xlabel('City')
ax.set_ylabel('Rail Line Count')

plt.xticks(rotation=90)
```
